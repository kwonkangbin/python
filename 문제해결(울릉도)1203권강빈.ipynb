#3.울릉도는 몇 월에 바람이 가장 강할까? 

```python
#df2['일시'].str[5:7]#월
#df2['일시'].str[7:]#일

df2['월'] = df2['일시'].str[5:7] #월속성 추가
```

```python
#월별로 묶기
re = df2.groupby('월')['평균풍속'].mean()
re.min()
```

```python
type(re)
```

```python
re2 = re.index[re == re.max()]
re2 = re.index[re == re.max()].tolist()
print(f'울릉도는 {re2[0]}월에 가장 바람이 강합니다.')
```

```python
# X 월
# Y 평균풍속
# 막대그래프 출려구 자료구조 변환 ->
y = re.tolist()
x = re.index.tolist()
plt.title('월별 풍속')
plt.bar(x,y,align = 'center', alpha = 0.5)
plt.xlabel('월')
plt.ylabel('풍속')

for i in range(0,len(x)):
  if re.max() == y[i]:
    plt.text(i,y[i],round(y[i],1),ha='center',color='r')
  else:
    plt.text(i,y[i],round(y[i],1),ha='center')
plt.show()
```
