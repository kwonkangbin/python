```python
fp = open('/content/drive/MyDrive/지역평균기온.txt','r',encoding = 'utf-8')
data = fp.readlines()
fp.close()
data
```

```python
dt = {} # 빈 딕셔너리

for i in data:
  i = i.replace("\n","")
  
  items = i.split(',')
  dt[items[0]] = int(items[1])

dt
  
```

```python
x = list(dt.keys())
y = list(dt.values())


plt.rc('font', family = 'NanumBarunGothic')
plt.title("지역평균기온")
plt.xlabel('도시')
plt.ylabel('온도')
for i in range(0,len(x)):
  plt.text(i,y[i],)
plt.show()
```
